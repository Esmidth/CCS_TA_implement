```python
tmp = int(len(tensor_array)*0.9)
ratio = [tmp,len(tensor_array)-tmp]
```

```python
test_loader = Data.DataLoader(
    dataset=test_db,
    batch_size=500,
    shuffle=True,
    num_workers=8
)
```
